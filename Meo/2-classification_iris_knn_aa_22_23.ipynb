# Classifiers comparison: decision trees and k-nearest neighbors on the dataset Iris


In the following program we compare the prediction results obtained by decision trees and k-nearest neighbors on the dataset Iris

The following cell shows the program training a decision tree and its results in preciction 

````python
from sklearn import tree 
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score # will be used to separate training and test

iris = load_iris()
clf = tree.DecisionTreeClassifier(criterion="entropy",random_state=300,min_samples_leaf=5,class_weight={0:1,1:1,2:1})
# clf = clf.fit(iris.data, iris.target)
# the above line is not needed as we will use cross_val_score and it will do the fitting for us
scores = cross_val_score(clf, iris.data, iris.target, cv=5) # score will be the accuracy
print(scores)
````

The following cell shows the training of k-nearest neighbors and its prediction results.
Here we use a uniform weighting setting (weights='uniform'): any neighbors weights the same in the majority voting aggregation.

````python
from sklearn import neighbors
n_neighbors = 11
clf_knn = neighbors.KNeighborsClassifier(n_neighbors, weights='uniform')
clf_knn = clf_knn.fit(iris.data, iris.target)
scores = cross_val_score(clf_knn, iris.data, iris.target, cv=5) # score will be the accuracy
print(scores)
# shows the model predictions  
for i in range(len(iris.target)):
    print(iris.data[i,:])
    # print(iris.data[i,:].reshape(1,-1))
    instance=iris.data[i,:].reshape(1,-1)
    # print(clf_knn.predict(instance))
    predicted=clf_knn.predict(instance)[0]
    print(predicted)
    if iris.target[i]==predicted:
        print(str(i)+" ok "+str(iris.target_names[iris.target[i]]))
    else:
        print(str(i)+" nok "+"true class: "+str(iris.target_names[iris.target[i]])+"; predicted: "+str(iris.target_names[predicted]))
````

In the following cell we use a varying weighting setting (weights='distance'): any neighbors weights inversely with its distance to the test instance in the majority voting aggregation.

````python
n_neighbors = 11
clf_knn2 = neighbors.KNeighborsClassifier(n_neighbors, weights='distance')
clf_knn2.fit(iris.data, iris.target)

for i in range(len(iris.target)):
    instance=(iris.data[i,:]).reshape(1, -1)
    predicted2=clf_knn2.predict(instance)[0]
    if iris.target[i]==predicted2:
        print(str(i)+" ok "+str(iris.target_names[iris.target[i]]))
    else:
        print(str(i)+" nok "+"true class: "+str(iris.target_names[iris.target[i]])+"; predicted: "+str(iris.target_names[predicted]))
print("Classification score of k-nn with distance weighting")
scores2 = cross_val_score(clf_knn2, iris.data, iris.target, cv=5,scoring='accuracy') # score will be the accuracy
print(scores2)
````

The following cell shows the tuning of the k-nn models with a varying value of k (number of nearest neighbors) and finds the best value of k (giving the maximum accuracy).

````python
import numpy as np
from numpy import zeros
from sklearn import neighbors
from sklearn.datasets import load_iris
iris = load_iris()
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4, random_state=0)

best_accuracy=0
best_k=1
A=np.zeros(len(y_train), dtype=float) # for storing the accuracy for each k
for n_neighbors in np.arange(1,len(y_train)+1):
    clf_knn3 = neighbors.KNeighborsClassifier(n_neighbors, weights='distance')
    # (n_neighbors=5, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=1, **kwargs)
    clf_knn3.fit(X_train, y_train)
    index=n_neighbors-1
    A[index]=clf_knn3.score(X_test, y_test)
    if best_accuracy<clf_knn3.score(X_test, y_test):
        best_accuracy=clf_knn3.score(X_test, y_test)
        best_k=n_neighbors
    print("k neighbors="+str(n_neighbors))
    print("accuracy="+str(clf_knn3.score(X_test, y_test)))
    
print("\n")
print("best k="+str(best_k))
print("best accuracy="+str(best_accuracy))

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

#plt.xticks(np.arange(1, len(y_train)+1, 8))
plt.yticks(np.arange(0.0,1.0,0.01))

plt.plot(np.arange(1,len(y_train)+1),A)

plt.show()
````

## 2. Mostrare uno scatter plot simile, ma ora il colore dei punti  determinato dalla classe stimata da k-nn, con un valore scelto di k a vostra scelta.

````python
# plot using the last two features the predicted points
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
knn = KNeighborsClassifier(n_neighbors=7, weights='distance')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
num_test_points = len(y_test)
print(f'Accuracy over {num_test_points} test points: {knn.score(X_test, y_test)}')
# plot the predicted points
plt.scatter(X_test[y_pred==0,2], X_test[y_pred==0,3], color='red', label='predicted setosa')
plt.scatter(X_test[y_pred==1,2], X_test[y_pred==1,3], color='blue', label='predicted versicolor')
plt.scatter(X_test[y_pred==2,2], X_test[y_pred==2,3], color='green', label='predicted virginica')
# plot the misclassified points
plt.scatter(X_test[y_pred!=y_test,2], X_test[y_pred!=y_test,3], color='black', label='misclassified', marker='x', s=100)
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.legend()
plt.title('Predicted classes using a k-nn classifier')
plt.show()
````

Come possiamo dal grafico sopra, usando un classificatore knn con ```k=7``` e ```weights='distance'``` si ottiene una accuracy del 96%. Infatti plottando i punti predetti si vede che vengono fatti solamente due errori di classificazione.

## 3. Visualizzate il ROC plot del migliore albero di decisione che avete addestrato nell'esercizio n.1.

````python
# Build the inflated model as it was the best one
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree

iris = load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

# inflating the training set by inflating the number of examples of class 1 (versicolor) by a factor of 10
inflated_X_train = np.concatenate((X_train, np.repeat(X_train[y_train==1], 10, axis=0)))
inflated_y_train = np.concatenate((y_train, np.repeat(y_train[y_train==1], 10, axis=0)))

inflated_tree = tree.DecisionTreeClassifier(criterion='entropy', random_state=0, min_samples_leaf=5)
inflated_tree.fit(inflated_X_train, inflated_y_train)

predicted_inflated = inflated_tree.predict(X_test)
print('Inflated tree result')
print(f'Accuracy over {len(y_test)} test points: {inflated_tree.score(X_test, y_test)}')
print(f'Number of misclassified points: {sum(predicted_inflated!=y_test)}')
print('-'*50)

# Use the previous k-nn classifier to predict the classes of the current test set
knn.fit(X_train, y_train)
predicted_knn = knn.predict(X_test)
print('K-nn result')
print(f'Accuracy over {len(y_test)} test points: {knn.score(X_test, y_test)}')
print(f'Number of misclassified points: {sum(predicted_knn!=y_test)}')
````

````python
from sklearn.metrics import roc_curve, auc

def plotROC(clf, name):
    probs = clf.predict_proba(X_test)
    c1 = probs[:, 0]
    c2 = probs[:, 1]
    c3 = probs[:, 2]
    # set to 1 if probability is greater than 0.5, 0 otherwise
    c1 = np.where(c1 > 0.5, 1, 0)
    c2 = np.where(c2 > 0.5, 1, 0)
    c3 = np.where(c3 > 0.5, 1, 0)

    fpr1, tpr1, thresholds1 = roc_curve(y_test, c1, pos_label=0)
    fpr2, tpr2, thresholds2 = roc_curve(y_test, c2, pos_label=1)
    fpr3, tpr3, thresholds3 = roc_curve(y_test, c3, pos_label=2)
    roc_auc1 = auc(fpr1, tpr1)
    roc_auc2 = auc(fpr2, tpr2)
    roc_auc3 = auc(fpr3, tpr3)

    plt.figure()
    plt.plot(fpr1, tpr1, color='red', lw=2, label='Setosa (area = %0.2f)' % roc_auc1)
    plt.plot(fpr2, tpr2, color='blue', lw=2, label='Versicolor (area = %0.2f)' % roc_auc2)
    plt.plot(fpr3, tpr3, color='green', lw=2, label='Virginica (area = %0.2f)' % roc_auc3)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC curve for {name}')
    plt.legend(loc="lower right")
    plt.show()

plotROC(inflated_tree, 'inflated tree')
plotROC(knn, 'k-nn')
````

## 4. Confrontare gli alberi decisione e k-nn sullo spazio ROC: per quali valori di (TPR,FPR) k-nn  meglio degli alberi di decisione?

````python
# I will use the code example from the scikit-learn documentation found here: https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html
# Adapted to the current situation
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import LabelBinarizer
import matplotlib.pyplot as plt
from sklearn.metrics import RocCurveDisplay

label_binarizer = LabelBinarizer().fit(y_train)
y_onehot_test = label_binarizer.transform(y_test)
inflated_score = inflated_tree.predict_proba(X_test)
knn_score = knn.predict_proba(X_test)

fig, ax = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Comparison of ROC curves for the two models', fontweight='bold')
for i in range(3):
    class_of_interest = i
    class_id = np.flatnonzero(label_binarizer.classes_ == class_of_interest)[0]
    ax = plt.subplot(1, 3, i+1)
    RocCurveDisplay.from_predictions(y_onehot_test[:, class_id], inflated_score[:, class_id], name='Inflated tree', ax=ax)
    RocCurveDisplay.from_predictions(y_onehot_test[:, class_id], knn_score[:, class_id], name='K-nn', ax=ax)
    ax.set_xlabel('False positive rate')
    ax.set_ylabel('True positive rate')
    ax.set_title(f'Class {iris.target_names[class_of_interest]}')
plt.show()
````

Come si pu vedere dal grafico sopra, k-nn  meglio del modello inflated tree per tutti i valori di (TPR,FPR).

## 5. Eseguite k-nn ma ora usate come funzione di distanza una funzione Kernel:

distance(x,y)= 1- k(x,y) 

dove k(x,y)  un Kernel Gaussian-like k(x,y) (per k(x,y) usate la Radial Basis Function con il parametro Gamma = 1/sigma^2) che controlla la sua ampiezza.

Il parametro gamma deve essere aggiustato (tuned) al valore ottimale, a secondo dell'accuratezza del k-nn (in modo simile a quanto abbiamo fatto precedentemente per il parametro n_neighbors). In questo caso, scegliete un valore di k=7.

````python
def kernel(x, y, sigma):
    '''
    Gaussian-like kernel function. Using Radial Basis Function (RBF) with gamma = 1/2*sigma^2 -- instead of 1/(sigma^2)
    '''
    gamma = 1/2*sigma**2
    return np.exp(-gamma * np.linalg.norm(x-y, ord=2)**2)

def distance(x, y, sigma):
    '''
    Distance function for the kernel k-nn classifier: distance = 1 - kernel(x, y)
    '''
    return 1 - kernel(x, y, sigma)
````

````python
# tune the sigma parameter
k = 7 # number of neighbors
sigmas = np.linspace(0, 40, 150)
accuracies = np.zeros(len(sigmas))
best_accuracy = 0
best_sigma = 0
for i in range(len(sigmas)):
    sigma = sigmas[i]
    knn = KNeighborsClassifier(n_neighbors=k, metric=distance, metric_params={'sigma': sigma})
    knn.fit(X_train, y_train)
    accuracies[i] = knn.score(X_test, y_test)
    if accuracies[i] > best_accuracy:
        best_accuracy = accuracies[i]
        best_sigma = sigma

print(f'Best accuracy = {best_accuracy} with sigma = {best_sigma:.2f}')

plt.figure()
plt.plot(sigmas, accuracies)
plt.xlabel('sigma')
plt.ylabel('accuracy')
plt.title('Accuracy over different values of sigma')
plt.show()
````

````python
# Use the best sigma to predict the classes of the current test set
knn = KNeighborsClassifier(n_neighbors=7, metric=distance, metric_params={'sigma': best_sigma})
knn.fit(X_train, y_train)
predicted_knn = knn.predict(X_test)
print(f'Kernel k-nn classifier with k = {k} and sigma = {best_sigma:.2f}')
print(f'Accuracy over {len(y_test)} test points: {knn.score(X_test, y_test)}')
print(f'Number of misclassified points: {sum(predicted_knn!=y_test)}')

# Plot the ROC curve for the kernel k-nn classifier
plotROC(knn, 'kernel k-nn')
````
